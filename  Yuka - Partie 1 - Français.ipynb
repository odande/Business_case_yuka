# Mission 4

Quel est le nombre de colonne appartenant à chaque quartile des taux de remplissage ? Autrement dit, combien de colonne sont remplies à 0%, entre 0 et 25%, entre 25 et 50%...

```
df_yuka.notnull().mean() * 100
```

```
dico_quartile = {
    '0' : 0,
    '0-25' : 0,
    '25-50' : 0,
    '50-75' : 0,
    '75-100' : 0
}

for colonne in df_yuka.columns:
  tr = df_yuka[colonne].notnull().mean() * 100
  if tr == 0:
    dico_quartile['0'] += 1
  elif 0 < tr <= 25:
    dico_quartile['0-25'] += 1
  elif 25 < tr <= 50:
    dico_quartile['25-50'] += 1
  elif 50 < tr <= 75:
    dico_quartile['50-75'] += 1
  else:
    dico_quartile['75-100'] += 1

dico_quartile
```

Un dataset trop pauvre en data n'est pas vraiment intéressant. Après avoir analyser les colonnes, nous nous sommes rendu compte que nous pouvons supprimer les colonnes avec moins de 60% de valeurs non nulles, sauf les colonnes pnns_groups_1 et pnns_groups_2, nutrition-score-fr_100g et nutrition-score-uk_100g. En effet ces 2 colonnes vont nous être utiles par la suite.

Si ce n'est pas déjà fait, supprime également les colonnes suivantes car elles ne sont pas vraiment utiles pour répondre à notre problématique :

```
df_yuka_fr_colagarder=['pnns_groups_1' , 'pnns_groups_2', 'nutrition-score-fr_100g' , 'nutrition-score-uk_100g']
df_yuka_fr_a_supprimer = ['creator', 'created_t', 'created_datetime', 'last_modified_t', 'brands_tags', 'countries', 'countries_tags', 'additives_tags', 'states', 'states_tags']

for colonne in df_yuka_fr.columns:
    if colonne not in df_yuka_fr_colagarder and df_yuka_fr[colonne].notnull().mean() < 0.60:
        df_yuka_fr.drop(colonne, axis=1, inplace=True)

df_yuka_fr.drop(columns=df_yuka_fr_a_supprimer, inplace=True, errors='ignore')



```
