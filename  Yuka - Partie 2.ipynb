```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```
df_yuka_fr=pd.read_csv('/content/yuka_fr.csv')
```

```
df_yuka_fr
```

```
sns.histplot(df_yuka_fr['nutrition-score-fr_100g'])
```

```
df_yuka_fr['nutrition-score-fr_100g'].hist(bins=50)
plt.title('Distribution du Nutrition Score FR')
plt.xlabel('Score')
plt.ylabel('Fréquence')
plt.show()

df_yuka_fr['nutrition-score-uk_100g'].hist(bins=50)
plt.title('Distribution du Nutrition Score UK')
plt.xlabel('Score')
plt.ylabel('Fréquence')
plt.show()
```

```
sns.heatmap(df_yuka_fr.select_dtypes('number'),cmap='crest',annot=True)
```

```
#Graph sur les PNNS du groupe 1
df_yuka_fr['pnns_groups_1'].value_counts().plot(kind='bar')
plt.title('Répartition des PNNS Groupes 1')
plt.ylabel('Nombre')
plt.show()
```

```
#Graph sur les PNNS du groupe 1
df_yuka_fr['pnns_groups_2'].value_counts().plot(kind='bar')
plt.title('Répartition des PNNS Groupes 2')
plt.ylabel('Nombre')
plt.show()
```

```
#Démonstration de la corrélation entre les nustriscore FR et UK pour 100g de produit et PNNS de groupe 1
sns.scatterplot(x='nutrition-score-fr_100g', y='nutrition-score-uk_100g', data=df_yuka_fr,hue='pnns_groups_1')
plt.title('Corrélation entre les Nutriscores FR et UK avec PNNS 1')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```

```
#Démonstration de la corrélation entre les nustriscore FR et UK pour 100g de produit et PNNS de groupe 2
sns.scatterplot(x='nutrition-score-fr_100g', y='nutrition-score-uk_100g', data=df_yuka_fr,hue='pnns_groups_2')
plt.title('Corrélation entre les Nutriscores FR et UK avec PNNS 2')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```

--CONCLUSION
- La base de donnée comporte beaucoup de zone non remplie notamment dans les colonnes PNNS 1 & PNNS 2, cela empèche une meilleure exploitation des données.
- Une corrélation existe entre les nutriscores et les produits.
- On constate, qu'une différence de régle sur les notations des nutriscores existent entre la France et le Royaume-Uni.
