```python
import requests
import folium
from shapely.geometry import shape

# Paramètres d'accès à la parcelle
code_insee = "35145"
section = "AB"
numero = "0346"

url = "https://apicarto.ign.fr/api/cadastre/parcelle"
params = {
    "code_insee": code_insee,
    "section": section,
    "numero": numero
}

# Récupération des données
response = requests.get(url, params=params)
data = response.json()

# Extraction de la géométrie (type MultiPolygon)
coords = data['features'][0]['geometry']['coordinates']

# On suppose ici un seul polygone (premier dans la liste)
polygone = coords[0][0]  # MultiPolygon > Polygone > Liste de points

# Calcul du centroïde pour centrer la carte
geom = shape(data['features'][0]['geometry'])
center = [geom.centroid.y, geom.centroid.x]

# Création de la carte
m = folium.Map(location=center, zoom_start=18)

# Ajout du polygone manuellement
folium.Polygon(
    locations=[(lat, lon) for lon, lat in polygone],
    color="blue",
    fill=True,
    fill_opacity=0.4,
    popup="Parcelle AB0346"
).add_to(m)

folium.LayerControl().add_to(m)
m
```
